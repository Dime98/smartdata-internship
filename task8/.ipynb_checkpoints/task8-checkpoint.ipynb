```python
def interquartile_range_method(df_col):
    sort_df_col = df_col.sort_values()
    q1 = sort_df_col.quantile(0.25)
    q3 = sort_df_col.quantile(0.75)
    iqr = q3 - q1
    upper_fence = q3 + (1.5 * iqr)
    lower_fence = q1 - (1.5 * iqr)
    outliers = ((sort_df_col < lower_fence) | (sort_df_col > upper_fence)).sum()
    # print(f"[{sort_df_col.name}] outliers: {outliers}")
    return outliers

def statistical_outlier_detection(df_col, get=False):
    data = df_col.to_numpy()
    mean = np.mean(data)
    sigma = np.std(data)     # standard deviation
    threshold = 6
    z = (data - mean)
    z /= sigma
    outliers = data[z > threshold]
    print(f"statistical_outlier [{df_col.name}] outliers: {len(outliers)}")
    if get == False:
        return len(outliers)
    else:
        return outliers
```

```python
statistical_outlier_detection(df['fixed acidity'])
statistical_outlier_detection(df['volatile acidity'])
statistical_outlier_detection(df['citric acid'])
statistical_outlier_detection(df['residual sugar'])
statistical_outlier_detection(df['chlorides'])
statistical_outlier_detection(df['free sulfur dioxide'])
statistical_outlier_detection(df['total sulfur dioxide'])
statistical_outlier_detection(df['density'])
statistical_outlier_detection(df['pH'])
statistical_outlier_detection(df['sulphates'])
statistical_outlier_detection(df['alcohol'])
statistical_outlier_detection(df['quality'])

# plt.scatter(df["alcohol"], df["alcohol"], label="alcohol")
# plt.scatter(df['fixed acidity'] , df['fixed acidity'], label = 'fixed acidity')
# plt.scatter(df['volatile acidity'] , df['volatile acidity'], label = 'volatile acidity')
# plt.scatter(df['citric acid'] , df['citric acid'], label = 'citric')
# plt.scatter(df['residual sugar'] , df['residual sugar'], label = 'residual')
# plt.scatter(df['chlorides'] , df['chlorides'], label = 'chlorides')
# plt.scatter(df['free sulfur dioxide'] , df['free sulfur dioxide'], label = 'free')
# plt.scatter(df['total sulfur dioxide'] , df['total sulfur dioxide'], label = 'total')
# plt.scatter(df['density'] , df['density'], label = 'density')
# plt.scatter(df['pH'] , df['pH'], label = 'pH')
# plt.scatter(df['sulphates'] , df['sulphates'], label = 'sulphates')
plt.scatter(df['alcohol'] , df['alcohol'], label = 'alcohol')
# plt.scatter(df['quality'] , df['quality'], label = 'quality')
plt.legend()
df["chlorides"].describe()
```

```python
def get_correlation(target_col, cols, sort=False):
    corr_dict = {}
    for i in cols:
        correlation = target_col.corr(df[i])
        corr_dict[i] = correlation

    if sort:
        corr_dict = dict(sorted(corr_dict.items(), key=lambda x: x[1], reverse=True))
    return corr_dict
```

```python
components_cols = [
    'fixed acidity', 
    'volatile acidity', 
    'citric acid', 
    'residual sugar', 
    'chlorides', 
    'free sulfur dioxide', 
    'total sulfur dioxide', 
    'density', 
    'pH', 
    'sulphates', 
]

target_col = df["alcohol"]
alcohol_correlation = get_correlation(target_col, components_cols, True)

target_col = df["quality"]
components_cols.append("alcohol")
quality_correlation = get_correlation(target_col, components_cols, True)
```
