```python
# 9. Write a program to create bar plot of scores by group and gender. Use multiple X 
# values on the same chart for men and women.
# Data:
# Sample Data:
# Means (men) = (22, 30, 35, 35, 26)
# Means (women) = (25, 32, 30, 35, 29)
men = [22, 30, 35, 35, 26]
women = [25, 32, 30, 35, 29]

# group_len = np.arange(len(men), len(women))
group_len = max([len(men), len(women)])
print(group_len)

x = np.arange(group_len)

width = 0.40
plt.bar(x-.2, men, width, label=men)
plt.bar(x+.2, women, width, label=women)

plt.legend()
```

```python
# 10. Write a program to add textures (black and white) to bars and wedges. Use the 
# data from previous exercise.

plt.bar(x*3+.25, men, label="men", color='black', edgecolor='black', hatch="/")
plt.bar(x*3-.25, women, label="women", color='white', edgecolor='black', hatch="//")

plt.legend()
```

```python
# 13. Write a program to represent the Poisson Probability for n values of X in range [0; 100].

from scipy.stats import poisson

x = np.arange(0, 100, 1)
y = poisson.pmf(x, mu=40, loc=10)
plt.plot(x, y)
```
