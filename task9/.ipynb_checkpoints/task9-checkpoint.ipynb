```python
def get_heat_map_corr(corr_df, figsize=(10, 9), remove_cols = [], title=None):

    df_temp = corr_df
    for i in remove_cols:
        df_temp = df_temp.drop(i,axis=1)

    corr = df_temp.corr()

    f, ax = plt.subplots(figsize=figsize)

    mask = np.triu(np.ones_like(corr, dtype=bool))
    cmap = sns.diverging_palette(230, 20, as_cmap=True)
    
    if title != None:
        plt.title(title)

    # sns.heatmap(corr, annot=True, cmap=cmap)
    sns.heatmap(corr, annot=True, mask = mask, cmap=cmap)
# get_heat_map_corr(red_df, remove_cols = ["type"], title="Red Wine Correlation Heat Map")
```

```python
def get_correlation(df, target_col, cols, sort=False):
    corr_dict = {}
    for i in cols:
        # print()
        # print(f"i          {i}")
        # print(f"target_col {target_col}")
        correlation = target_col.corr(df[i])
        corr_dict[i] = correlation

    if sort:
        corr_dict = dict(sorted(corr_dict.items(), key=lambda x: x[1], reverse=True))
    return corr_dict

# =======================================

# target_col = red_df['alcohol']

# cols = [c for c in red_df.columns]
# excl_cols = ['type', "alcohol", "quality"]
# cols = [c for c in red_df.columns if c not in excl_cols]
# # cols

# red_corr = get_correlation(red_df, target_col, cols, sort=True)
# red_corr
```

### correlation of components against 'alcohol' of red and white wine seprately

```python
target_col = red_df['alcohol']

cols = [c for c in red_df.columns]
excl_cols = ['type', "alcohol", "quality"]
cols = [c for c in red_df.columns if c not in excl_cols]

red_corr = get_correlation(red_df, target_col, cols, sort=True)
red_corr

print(f"Red wine {target_col.name} correlation:")
for i in list(red_corr.keys())[:4]:
    print(i, red_corr[i])
```

```python
target_col = white_df['alcohol']

cols = [c for c in white_df.columns]
excl_cols = ['type', "alcohol", "quality"]
cols = [c for c in white_df.columns if c not in excl_cols]

white_corr = get_correlation(white_df, target_col, cols, sort=True)

print(f"white wine {target_col.name} correlation:")
for i in list(white_corr.keys())[:4]:
    print(i, white_corr[i])
```
