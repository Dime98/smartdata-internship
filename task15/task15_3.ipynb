```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
# df contains prices of top 20 most common dog breeds and least common 10
df = pd.read_csv('dog_prices_top20_bottom10.csv')
df2 = pd.read_csv("20170308hundehalter.csv")
```

```python
df=df.drop('price',axis=1)
```

```python
# drop 2 rows where i didn't find price for dogs
df.dropna(how='any', inplace=True)
```

```python
# get number of dogs per breed
df2 = df2['RASSE1'].value_counts().sort_values(ascending=False)
df2
# df2.index
```

```python
# delete dog breeds that are not in dog prices csv
for i in df2.index:
    # print(i)
    if i not in df['breed'].values:
        df2.drop(i,inplace=True)
df2
```

```python
df[['from', 'to']] = df[['from', 'to']].astype(int)
df.head()
```

```python
df['mean_price'] = (df['from'] + df['to']) / 2
df
```

```python
# add numbers of dogs to df with prices

df['nr_of_dogs'] = None
df
for i in range(df.shape[0]):
    # print()
    df_breed_name = df.iloc[i].values[0]
    # print(f"df  {df_breed_name}")
    # print(f"df  {df.iloc[i]['nr_of_dogs']}")
    # print(f"df2 {(df2.loc[df_breed_name])}")
    df.iloc[i,df.columns.get_loc('nr_of_dogs')] = df2.loc[df_breed_name]
    # print(f"df  {df.iloc[i]['nr_of_dogs']}")
# df2
```

```python
df
```

```python
# if high number of dogs represent high demand
df.sort_values(['nr_of_dogs', 'mean_price'], ascending=[False,False])
```

```python
# if high number of dogs represent oversupply, (prices could drop)
df.sort_values(['nr_of_dogs', 'mean_price'], ascending=[True,True])
```
